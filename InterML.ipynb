```python
from sklearn.ensemble import HistGradientBoostingClassifier
```

```python
traindf = pd.read_csv("train.csv")

train_Y = traindf["label"]
train_X = traindf.drop("label", axis=1)

print(train_Y)
```

```python
knn = HistGradientBoostingClassifier()
knn.fit(train_X, train_Y)
```

```python

testdf = pd.read_csv("test.csv")
test_Y = testdf["label"]
test_X = testdf.drop("label", axis=1)
answerdf = knn.predict(test_X)
print(answerdf)
```

```python
c = 1
modedanswers = []
tempL = []
for x in answerdf:
    tempL.append(x)
    if c%5==0:
        modedanswers.append(statistics.mode(tempL))
        c = 0
        tempL = []
    c+=1

print(modedanswers)
        

```

```python
i = 0
correct = 0
for j in answerdf:
    if j == test_Y[i]:
        correct += 1
    i+=1
print(correct/i)
```
