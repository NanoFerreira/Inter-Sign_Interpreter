```python
#library importingg

import pandas as pd
import numpy as np
from PIL import Image
```

```python
# Opening up the ASL img file and turning it into an array.

#user_img = input("Type the image's address: ")
file = Image.open('asadf.jpg')
new_f = file.resize((28,28))
new_f.save('downscaled.png')
numArray = np.asarray(new_f)
```

```python
print(len(numArray[0]))
```

```python
# If the image is not greyscale, we greyscale it

newNumArr = []
if(numArray.ndim>2):
    for i in numArray:
        row = []
        for j in i:
            sum = 0
            for k in range (len(j)-1):
                sum+=j[k]
            row.append(sum//(len(j)-1))
        newNumArr.append(row)

numArray = newNumArr
```

```python
#Now we're taking the resulting greyscaled Array and making it 1D to fit train data

d1Arr = []

for i in numArray:
    for j in i:
        d1Arr.append(j)

numArray = d1Arr        
```

```python
for x in range(len(d1Arr)):
    if(x%28==0):
        print()
    print(d1Arr[x], end=", ")  
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
traindf = pd.read_csv("sign_mnist_train.csv")

train_Y = traindf["label"]
train_X = traindf.drop("label", axis=1)
```

```python
knn =MLPClassifier()
knn.fit(train_X, train_Y)
```

```python
testdf = pd.read_csv("sign_mnist_test.csv")
test_Y = testdf["label"]
test_X = testdf.drop("label", axis=1)
answerdf = knn.predict(test_X)
print(answerdf)
```
